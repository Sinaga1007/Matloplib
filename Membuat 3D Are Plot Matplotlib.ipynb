```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Sample Dataset

n = np.linspace(-3.0, 3.0, 100)

x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

#Akan merepresentasikan koordinat
print(x.shape)
print(y.shape)
print(z.shape)
```

```python
#3D Wireframe

fig = plt.figure()                        #Membuat objek Figure
ax = plt.axes(projection='3d')    

ax.plot_wireframe(x, y, z, color='yellow') #Membutuhkan 3 parameter, warna optional

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Wireframe')

plt.show()
```

```python
#3D Surface Plot

fig = plt.figure()                         #Membuat objek Figure
ax = plt.axes(projection='3d')    

#ax.plot_surface(x, y, z, color='yellow')   #Membutuhkan 3 parameter, warna optional
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')

plt.show()
```
