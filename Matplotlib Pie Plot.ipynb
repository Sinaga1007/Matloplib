```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

```python
#Simple Pie Plot

kategori = ['A','B','C','D','E']
data = [220, 150, 270, 230, 250]
```

```python
#Untuk melakukan plotting

plt.pie(data,             #Nilai Integer
        labels=kategori,  #label sifatnya optional
        autopct='1.1f%%', #Berguna untuk menampilkan persentase dari setiap area pada pie plot dan bersifat optional
        startangle=80)    #Berguna untuk rotasi pada pie plot

plt.title('Contoh Pie Plot') #Menampilkan judul
plt.show()
```

```python
#Pengaturan Legend dan Color

kategori = ['A','B','C','D','E']
data = [220, 150, 270, 230, 250]
warna = ['pink', 'cyan', '#e83364', 'yellowgreen', 'skyblue']
```

```python
plt.pie(data,             #Nilai Integer
        labels=kategori,  #label sifatnya optional
        colors=warna,     #Pengaturan warna
        autopct='1.1f%%', #Berguna untuk menampilkan persentase dari setiap area pada pie plot dan bersifat optional
        startangle=80)    #Berguna untuk rotasi pada pie plot

plt.legend() #Keterangan plotting

plt.title('Contoh Pie Plot') #Menampilkan judul
plt.show()
```

```python
#Exploded Pie Plot

kategori = ['A','B','C','D','E']
data = [220, 150, 270, 230, 250]
warna = ['pink', 'cyan', '#e83364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08] #akan berkolerasi dengan kategori
```

```python
plt.pie(data,                #Nilai Integer
        labels=kategori,     #label sifatnya optional
        colors=warna,        #Pengaturan warna
        explode=explode_var, #Pengaturan untuk potongan pada pie plot
        autopct='1.1f%%',    #Berguna untuk menampilkan persentase dari setiap area pada pie plot dan bersifat optional
        startangle=80)       #Berguna untuk rotasi pada pie plot

plt.legend() #Keterangan plotting

plt.title('Contoh Pie Plot') #Menampilkan judul
plt.show()
```
