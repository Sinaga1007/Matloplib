```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Sample Dataset

#loc untuk mengidentifikasi nilai mean atau rata" dari distribusi normal yg dihasilkan
#scale ini setara dengan standar deviation

np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

```python
#Simple Violin Plot

plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Pengaturan Pada Violin Plot

plt.violinplot(data,        
               showextrema=False,   #Digunakan utk menampilkan garis tengah, serta mengidentifikasi nilai max dan min dari dataset
               showmeans=False,     #Digunakan utk menampilkan nilai rata-rata     
               showmedians=False,   #Digunakan utk menampilkan nilai tengah
               quantiles=None)      #Akan menerima nilai berupa colection data types

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Horizontal Violin Plot

plt.violinplot(data, vert=False)

plt.title('Horizontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Multiple Violin Plot

np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(70, 20, 200)
data3 = np.random.normal(80, 30, 200)
data4 = np.random.normal(50, 25, 200)

data = [data1, data2, data3, data4]
```

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
