```python
import matplotlib.pyplot as plt ####Modul utama dalam floating data
import numpy as np
```

```python
##Mempersiapkan sampel data
x = [1, 2, 3, 4]
y = [1, 3, 4, 2]

fig, ax = plt.subplots() ####Membuat sebuah figure dan sebuah axes
ax.plot(x, y) ####Untuk melakukan plotting pada sebuah axes
```

```python
####Alternatif lain dalam memplotting data
x = [1, 2, 3, 4]
y = [1, 3, 4, 2]

plt.plot(x, y)
####Hasil visualisasi datanya sama
```

```python
# Cara membuat figure dan axes

#Model 1
fig = plt.figure() ####Hanya membuat 1 objek figure tanpa disertai axes
```

```python
#Model 2
fig.ax = plt.subplots() ####Membuat sebuah figure dan sebuah axes
```

```python
#Model 3

fig, axs = plt.subplots(3, 2) ####Akan menghasilkan sebuah figure dengan 3x2 grid axes
```

```python
#Mengenal Object Oriented Style untuk melakukan ploting

x = np.linspace(0, 2, 100) ##Yg ini digunakan untuk menghasilkan sejumlah deret bilangan
x
```

```python
fig, ax = plt.subplots() #Pemangilan method plt.subplots yang tidak ada parameter akan menghasilkan sebauh figure dan sebuah axes

####Memplotting tiga variasi data pada axes
ax.plot(x, x, label='linear') ##Menghasilkan garis lurus, karena nilai x dan y nya sama
ax.plot(x, x**2, label='quadratic') ## nilai y diperoleh dari x**2
ax.plot(x, x**3, label='cubic') ##nilai y merupakan hasil dari x**3

ax.set_xlabel('x label') #Digunakan untuk menyertakan x-label pada axes
ax.set_ylabel('y label') #Digunakan untuk menyertakan y-label pada axes
ax.set_title("Simple Plot") ##Untuk menyertakan title pada axes
ax.legend() ##Untuk menyertakan legend
```

```python
x = np.linspace(0, 2, 100)
x
```

```python
##Dibagian ini kita tidak perlu membuat objek figure dan axes
plt.plot(x, x, label='linear') ##Menghasilkan garis lurus, karena nilai x dan y nya sama
plt.plot(x, x**2, label='quadratic') ## nilai y diperoleh dari x**2
plt.plot(x, x**3, label='cubic') ##nilai y merupakan hasil dari x**3

plt.xlabel('Label x') #Digunakan untuk menyertakan Label x pada plt
plt.xlabel('Label y') #Digunakan untuk menyertakan Label y pada plt
plt.title('Simple Plot') ##Untuk menyertakan title/Judul pada plt
plt.legend() ##Untuk menyertakan legend

#Hasilnya tetap sama hanya membedakan pada Pemanggilan method plot bukan pada objek axes melainkan plt
```
