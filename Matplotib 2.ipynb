```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Membuat plotting sederhana

plt.plot([3, 6, 7, 12])
plt.ylabel('Sumbu Y')
plt.show() #Fungsinya untuk tidak menampilkan tulisan
```

```python
#Membuat 2 deret bilangan

plt.plot([3, 4, 6, 9], [4, 2, 7, 15])
plt.show()
```

```python
#Pengaturan Format Pada Plot

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'g*') #2 deret bilangan dan parameter ketiga sebagai pengatur bentuk marker
plt.axis([0, 6, 0, 20]) #fungsi axis untuk menyertakan list yaitu berupa (xmin,xmax,ymin,ymax)
plt.show()
```

```python
#Multi Plot dalam satu pemanggilan fungsi plot()

t = np.arange(0., 5., 0.2) #0 untuk nilai minimum, 5 maximum dan 0.2 sebagai interval
t
```

```python
plt.plot(t, t, 'r--') #r-- berbentuk garis putus-putus dan bewarna merah
plt.plot(t, t**2, 'bs') #bs berbentuk square dan bewarna biru
plt.plot(t, t**3, 'g^') #g^ berbentuk segitiga dan bewarna hijau
plt.show()
```

```python
#Cara lain untuk Melakukan multi plot dalam satu pemanggilan fungsi plot()

plt.plot(t, t, 'r--',
         t, t**2, 'bs',
         t, t**3, 'g^')
plt.show()
```

```python
#Plotting dengan keywords

data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50), #Melakukan pengacakan bilangan integer dengan jangkauan 0-50 sebanyak 50 buah
        'd': np.random.randn(50)} #melakukan pengacakan bilangan integer sebanyak 50 bilangan
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100 

data
```

```python
plt.scatter('a', 'b', data=data) #Parameter data ini diambil dari data
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
#Representasinya berupa warna
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
#Plotting untuk tipe data categorical

names = ['group_a', 'group_b', 'group_c'] #kategori atau label
values = [1, 10, 100] #Tipe data numerical
```

```python
plt.plot(names, values)
plt.show()
```

```python
plt.scatter(names, values)
plt.show()
```
