```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
x = np.arange(0.0, 2.0, 0.01) #Parameter ketiga berfungsi untuk menentukan interval nilainya 
x
```

```python
s = np.sin(2 * np.pi * x) #Untuk menampung nilai hasil operasi sin dari dataset yang ditampung variabel x  
s
```

```python
fig, ax = plt.subplots() #akan menghasilkan 2 buah nilai yang perlu kita tampung dari 2 variabel 

ax.plot(x, s) #Melakukan plotting
ax.set(xlabel='Nilai x',
       ylabel='Nilai sin(x)',
       title='Visualisasi nilai sin')
ax.grid() #Untuk memunculkan grid

plt.show()
```

```python
#Multiple Subplots

x1 = np.linspace(0.0, 4.0, 100) #parameter 1 digunakan untuk mengatur nilai terkecil, parameter 3 berfungsi untuk jumlah data point yg digenerate
x2 = np.linspace(0.0, 3.0, 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
x1
```

```python
#Multiple Subplots dengan OO Style

fig, (ax1, ax2) = plt.subplots(2, 1) #2 merepresentasikan jmlh baris, 1 merepresentasikan jmlh kolom
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai Cos(x1)')

ax2.plot(x1, y1, 'g.-')
ax2.set_ylabel('Nilai Cos(x2)')

ax2.set_xlabel('Nilai x')

plt.show()
```

```python
#Multiple Subplots dengan pyplot Style

plt.subplot(2, 1, 1) #2 parameter utama utk mengatur layout, Paramater ketiga berfungsi untuk mengendalikan akses pertama 
plt.plot(x1, y1, 'ro-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai $cos(x_1)$')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'g.-')
plt.xlabel('Nilai $cos(x_2)$')
plt.ylabel('Nilai $x$')

plt.show()
```
