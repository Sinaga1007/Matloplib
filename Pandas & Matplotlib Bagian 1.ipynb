```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print(matplotlib.__version__)
print(pd.__version__)
```

```python
#Sample Dataset

df = pd.DataFrame({
     'Nama':['Bima', 'Catherine', 'Nono', 'Bono', 'Akang', 'Wina', 'Mamat'],
     'Usia':[22, 65, 27, 18, 55, 32, 21],
     'Gender':['L', 'P', 'L', 'L', 'L', 'P', 'L'],
     'Provinsi':['Medan', 'Sulut', 'Medan', 'Sulut', 'Medan', 'Papua', 'Papua'],
     'n_anak':[3, 1, 0, 4, 2, 1, 6],
     'n_pets':[4, 0, 1, 5, 3, 3, 2]
})

df
```

```python
#Scatter Plot

df.plot(kind='scatter', x='n_anak', y='n_pets', color='red') #kind utk menentukan jenis plotting yang dimunculkan, color utk menentukan warna marker

plt.title('Scatter Plot')
plt.xlabel('Jumlah Anak')
plt.ylabel('Jumlah Pets')

plt.show()
```

```python
#Histogram

df[['Usia']].plot(kind='hist', bins=[0, 20, 40, 60, 80, 100]) #kind utk menentukan jenis plotting yang dimunculkan

plt.title('Histogram')
plt.xlabel('Usia')
plt.ylabel('Jumlah')

plt.show()
```

```python
#Line Plot

df.plot(kind='line', x='Nama', y='n_anak', color='orange')

plt.title('Line Plot')
plt.ylabel('Jumlah Anak')

plt.show()
```

```python
#Multiple Line Plot

df.plot(kind='line', x='Nama', y='n_anak', color='black')
df.plot(kind='line', x='Nama', y='n_pets', color='green')

plt.title('Line Plot')
plt.ylabel('Jumlah')

plt.show()
```

```python
#Menyatukan line plot

ax = plt.gca() #Fungsinya akan mengembalikan axes

df.plot(kind='line', x='Nama', y='n_anak', color='black', ax=ax)
df.plot(kind='line', x='Nama', y='n_pets', color='green', ax=ax)

plt.title('Line Plot')
plt.ylabel('Jumlah')

plt.show()
```

```python
#Bar Plot

df.plot(kind='bar', x='Nama', y='Usia')

plt.title('Bar Plot')
plt.ylabel('Usia')

plt.show()
```

```python
df.groupby('Provinsi')['Nama'].count().plot(kind='bar')

plt.title('Bar Plot')
plt.ylabel('Jumlah')

plt.show()
```

```python
#Stacked bar plot dengan dua level group by

df.groupby(['Provinsi', 'Gender']).size().unstack().plot(kind='bar', stacked=True)

plt.title('Bar Plot')
plt.ylabel('Jumlah')

plt.show()
```

```python
df.groupby(['Gender', 'Provinsi']).size().unstack().plot(kind='bar', stacked=True) #Melakukan proses kalkulasi dgn memanggil fungsi .size dan unstack

plt.title('Stacked Bar Plot')
plt.ylabel('Jumlah')

plt.show()
```
