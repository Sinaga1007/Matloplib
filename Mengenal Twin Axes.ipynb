```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Kasus Twin Axes 1

x = np.linspace(1, 10, 25)  #2 parameter akan menentukan nilai terkecil dan terbesar, parameter 3 utk jumlah poin dari linspace
x
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, np.exp(x), 'bs-', label='exp')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx() #Pemanggilan twinx akan menghasilkan akses baru yg akan ditempatkan diatas akses yg pertama 
ax2.plot(x, np.log(x), 'ro-', label='log')
ax2.set_ylabel('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc='upper left')

plt.show()
```

```python
#Kasus Twin Axes 2

x = np.random.randint(16, 40, size=30)
x
```

```python
def C2F(celsius=0):                  #Akan menerima sebuah nilai berupa celsius dan dikonversi ke fahrenheit
    return (celsius * 1.8) + 32

def konversi_sumbu(ax1):
    y1, y2 = ax1.get_ylim()
    ax2.set_ylim(C2F(y1), C2F(y2))   #Konversi menjadi Fahrenheit
    ax2.figure.canvas.draw()         #Untuk menggambar canvas
    
fig, ax1 = plt.subplots()            #Plotting data
ax2 = ax1.twinx()

ax1.callbacks.connect('ylim_changed',konversi_sumbu)
ax1.plot(x)

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celsius')
ax2.set_ylabel('Fahrenheit')

fig.suptitle('Temperatur Udara')
plt.show()
```
