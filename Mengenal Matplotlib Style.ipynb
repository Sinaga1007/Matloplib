```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Fungsi Plotting

def gambar_barplot():
    data1 = [30, 85, 90, 50, 70]
    data2 = [45, 30, 25, 60, 15]
    kategori = ['A', 'B', 'C', 'D', 'E']
    
    x = np.arange(len(kategori))
    
    width = 0.35
    
    plt.bar(x-width/2, data1, width, label='Data1')
    plt.bar(x+width/2, data2, width, label='Data2')
    
    plt.xticks(x, kategori)
    
    plt.grid(linestyle='--',
             linewidth=1,
             axis='y',
             alpha=0.75)
    
    plt.xlabel('Kategori')
    plt.ylabel('Jumlah')
    plt.title('Contoh Grouped Bar Plot')
    
    plt.legend()
    plt.show()
```

```python
gambar_barplot()
```

```python
#Matplotlib Style

plt.style.available #Untuk mengetahui Daftar Style yang dimiliki oleh matplotlib
```

```python
#Cara untuk mengubah style matplotlib

style.use('grayscale')
gambar_barplot()
```
