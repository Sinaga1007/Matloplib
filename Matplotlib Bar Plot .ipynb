```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Simple Bar Plot

data = [30, 40, 65, 130, 235] #Deret bilangan
kategori = ['A','B','C','D','E'] #Deret Karakter
```

```python
#Memvisualisasikan Data

plt.bar(kategori, data) #Untuk menghasilkan bar plot
plt.grid() #Untuk pemanggilan grid pada bar plot

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
#Pengaturan Grid dan Color

data = [30, 40, 65, 130, 235] #Deret bilangan
kategori = ['A','B','C','D','E'] #Deret Karakter
```

```python
#plt bar digunakan untuk mengganti warna pada bar plot

plt.bar(kategori,
        data,
        color='green', #warna
        alpha=0.50) #mengukur tingkat transparansi pada bar plot

plt.grid(linestyle='--',
         linewidth=2, #Ketebalan line nya
         axis='x', #pilihan x, y, both
         alpha=0.50) #mengukur transparansi pada garis di bar plot

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
#Horizontal Bar Plot

data = [30, 40, 65, 130, 235] #Deret bilangan
kategori = ['A','B','C','D','E'] #Deret Karakter

#plt.barh(kategori, data) #plt barh digunakan untuk membuat horizontal pada bar plot
plt.barh(kategori[::-1], data[::-1]) #untuk membalikan urutan pada bar plot

plt.grid(linestyle='--',
         linewidth=2, #Ketebalan line nya
         axis='x', #pilihan x, y, both
         alpha=0.65) #mengukur transparansi pada garis di bar plot

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
#Stacked Bar Plot

data1 = [30, 90, 75, 45, 65]
data2 = [45, 35, 25, 60, 20]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori, data1, label='Data 1')
# plt.bar(kategori, data2, label='Data 2')

plt.bar(kategori, data2, label='Data 2', bottom=data1) #parameter bottom digunakan untuk menempatkan diatas pada bar plot

plt.legend() #Menampilkan keterangan label

plt.grid(linestyle='--',
         linewidth=1, #Ketebalan line nya
         axis='y', #pilihan x, y, both
         alpha=0.65) #mengukur transparansi pada garis di bar plot

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
#Grouped Bar Plot

data1 = [30, 90, 75, 45, 65]
data2 = [45, 35, 25, 60, 20]
kategori = ['A','B','C','D','E']

x = np.arange(len(kategori)) #untuk menghasilkan deret bilangan dari kategori
x
```

```python
width = 0.35 #Mengukur tipisnya bar plot

plt.bar (x-width/2, data1, width, label='Data 1') #Parameter 1 akan berasosiasi dengan nilai index yang diatas
plt.bar (x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)

plt.grid(linestyle='--',
         linewidth=1, #Ketebalan line nya
         axis='y', #pilihan x, y, both
         alpha=0.65) #mengukur transparansi pada garis di bar plot

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.legend()
plt.show()
```
