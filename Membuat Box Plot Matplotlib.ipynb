```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Sample Dataset

np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200) #Akan menghasilkan sekumpulan bilangan acak yg akan membentuk distribusi normal
data
```

```python
#Simple Box Plot

plt.boxplot(data, labels=['Data'])  #Parameter untuk membuat boxplot

plt.title('Simple Box Plot')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

```python
#Notched Box Plot

plt.boxplot(data, labels=['Data'], notch=True)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Pengaturan marker untuk Outlier

plt.boxplot(data,
            labels=['Data'],
            showfliers=True, #Untuk mengatur outlier
            flierprops={'markerfacecolor':'r','marker':'*'}) #Untuk mengatur warna dan bentuk outlier

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu Y')
          
plt.show()          
```

```python
#Horizontal Box Plot

plt.boxplot(data, labels=['Data'], vert=False) #Bisa diatur hori ataupun ver dengan parameter vert

plt.title('Notched Box Plot')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Multiple Box Plot

np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(70, 20, 200)
data3 = np.random.normal(80, 30, 200)
data4 = np.random.normal(50, 25, 200)

data = [data1, data2, data3, data4]
labels = ['Data 1', 'Data 2', 'Data 3', 'Data 4']
```

```python
plt.boxplot(data, labels=labels)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu Y')

plt.show()
```
