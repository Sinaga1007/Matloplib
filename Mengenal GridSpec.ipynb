```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Simple 2x2 Layout dengan subplots

siswa = ['Agung', 'Jaka', 'Nia', 'Tina']
uas = [62, 84, 95, 73]

fig, axes = plt.subplots(ncols=2, nrows=2, constrained_layout=True)  #Fungsi constrained layout ketika true utk menghindari tumpang tindih

axes[0][0].bar(siswa, uas)
axes[0][1].pie(uas, labels=siswa)
axes[1][0].pie(uas, labels=siswa, explode=[0, 0, 0, 2])
axes[1][1].barh(siswa, uas)

plt.show()
```

```python
#Simple 2x2 Layout dengan GridSpec

fig = plt.figure(constrained_layout=True)

gs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)  #objek gridspecnya dibentuk dengan memanggil fungsi gridspec.GridSpec

ax1 = fig.add_subplot(gs[0, 0])  #gridspec akan memposisikan layout yang kita buat, cth: baris ke 0, colom ke 0
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 2])
ax4.barh(siswa, uas)

plt.show()
```

```python
fig = plt.figure(constrained_layout=True)

gs = fig.add_gridspec(ncols=2, nrows=2) #objek gridspecnya dibentuk dengan memanggil fungsi fig.add_gridspec

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 2])
ax4.barh(siswa, uas)

plt.show()
```

```python
#Rows Span dan Cols Span dengan GridSpec

fig = plt.figure(constrained_layout=True)

gs = fig.add_gridspec(ncols=2, nrows=2)

ax1 = fig.add_subplot(gs[0, :]) #Memakai blank titik dua blank artinya kita mau menggunakan keseluruhan kolom yang tersedia
# ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
# ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 2])
ax4.barh(siswa, uas)

plt.show()
```

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(ncols=3, nrows=3)

ax1 = fig.add_subplot(gs[0, :])
ax1.set_title('gs[0 :]')

ax2 = fig.add_subplot(gs[1, :-1])
ax2.set_title('gs[1 :-1]')
ax2.barh(siswa, uas)

ax3 = fig.add_subplot(gs[1:, -1])
ax3.set_title('gs[1:, -1]')
ax3.bar(siswa, uas)

ax4 = fig.add_subplot(gs[-1, 0])
ax4.set_title('gs[-1, 0]')

ax5 = fig.add_subplot(gs[-1, -2])
ax5.set_title('gs[-1, -2]')

plt.show()
```
